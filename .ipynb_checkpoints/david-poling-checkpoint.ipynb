```python

import pandas as pd
```

```python
#read in dataset
arrest_data = pd.read_csv("arrest-data.csv")
```

```python
#drop columns that are not relevant to this analysis
arrest_data = arrest_data.drop(columns = ["PK","CCR","INCIDENTZONE","INCIDENTTRACT","COUNCIL_DISTRICT","PUBLIC_WORKS_DIVISION","X","Y"])
```

```python
#view data we are working with
arrest_data.head()
#arrest_data.size
#list(arrest_data["OFFENSES"].unique())
```

```python
by_neighborhood = (arrest_data.drop(columns=["AGE","GENDER","RACE","ARRESTTIME","ARRESTLOCATION","OFFENSES","INCIDENTLOCATION"]))
```

```python
counts = by_neighborhood.value_counts()
counts[0]
```

```python
counts.plot(kind = "barh", y = counts, figsize = (20,20))
```

```python

from plotnine import ggplot, geom_point, aes, stat_smooth, facet_wrap, geom_bar
```

```python
import datetime
```

```python
for index, row in arrest_data.iterrows():
    year = int(row["ARRESTTIME"][:4])
    month = int(row["ARRESTTIME"][5:7])
    day = int(row["ARRESTTIME"][8:10])
    newdate = datetime.datetime(year,1,1)
    print(newdate)
    
    arrest_data.loc[index,"ARRESTTIME"] = year


```

```python

sorted_arrests_by_date = arrest_data.sort_values(["ARRESTTIME"], ascending=False)
sorted_arrests_by_date.head()
```

```python
sorted_arrests_by_date = sorted_arrests_by_date.groupby(by=["ARRESTTIME","INCIDENTNEIGHBORHOOD"]).size()
```

```python
sorted_arrests_by_date.head(100)
```

```python
test_test = sorted_arrests_by_date.reset_index()
test_test.head(100)
test_test["count"] = test_test[0]
test_test.head(100)
```

```python
ggplot(test_test,aes(x="ARRESTTIME",y="count"))+geom_point(aes(color="INCIDENTNEIGHBORHOOD"))
```
