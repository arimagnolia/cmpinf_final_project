```python
import graphlib
import pandas as pd
import matplotlib.pyplot as plt

file = pd.read_csv("median_age_of_death.csv", index_col = 'NEIGHBORHOOD').dropna() 

file = file.sort_values(by = "TOTAL MD AGE AT DEATH", ascending=True)

file.plot(kind='barh', y='TOTAL MD AGE AT DEATH', figsize=(10,file.shape[0] * 0.3), title="Median Age Of Death by Neighborhood")


```
